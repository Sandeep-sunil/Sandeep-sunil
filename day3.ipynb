```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
n=int(input("Enter the number of students:"))
hours =[]
scores =[]
for i in range(0,n):
  a=int(input("Enter study hrs: "))
  b=int(input("Enter score: "))
  hours.append(a)
  scores.append(b)
sort_hrs=sorted(hours)
sort_scores=sorted(scores)
print(sort_hrs)
print(sort_scores)
x = np.array(sort_hrs).reshape(-1,1)
y = np.array(sort_scores)
z=LinearRegression()
z.fit(x,y)
m=int(input("Enter no.of students: "))
new_hours=[]
for i in range(0,m):
  f=int(input("Enter study hrs: "))
  new_hours.append(f)
sort_new=sorted(new_hours)
x_new=np.array(sort_new).reshape(-1,1)
new_scores=z.predict(x_new)
print(sort_new)
print(new_scores)
```
